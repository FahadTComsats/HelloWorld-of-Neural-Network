We have create a simple model with only one dence layer and only one neuron. 

```python
model = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape= [1])])
```

For the model part. I have used stochastic gradient descent optimizer and for the loss function, I have used mean squared error.

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

To feed some data to the model, I have created some data.

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

Give the data to the model and included 500 epochs

```python
model.fit(xs,ys,epochs=500)
```

As you can see. Our model starts out with the pretty high loss but eventually comes down as the number of epochs increases

```python
print(model.predict([10.0]))
```
